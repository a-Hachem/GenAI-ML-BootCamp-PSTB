Vectorization with Sentence Transformers

```python
import torch
from sentence_transformers import InputExample   # Required for creating sentence embedding examples
```

```python
def example_create_fn(doc1: pd.Series) -> InputExample:
    """
    Converts a row from the DataFrame into a format suitable for SentenceTransformer.

    Parameters:
    - doc1 : (pd.Series) A single row from the DataFrame (contains 'id' and 'title').

    Returns:
    - An InputExample object containing:
      - guid (string): The unique identifier (ID) of the document.
      - texts (list of str): The title of the document as a list (required format for SentenceTransformer).
      - label (float): A placeholder label (not used here, so we set it to 0.0).
    """
    return InputExample(guid=str(doc1.id),  # Convert the ID to a string (required format)
                        texts=[doc1["title"]],  # Wrap the title in a list (expected format)
                        label=0.0)  # Set a placeholder label (not used for unsupervised embeddings)
```

#  Exercise 3:

FAISS Indexing and Search

```python
import numpy as np
import faiss
import pandas as pd
from sentence_transformers import SentenceTransformer

# # Charger le modèle
# model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

# Assume `pdf_subset` contains "id" and "title" columns, create a copy to prevent modifications
pdf_to_index = pdf_subset.copy()

# Convert the "title" column into a list of titles
titles_list = pdf_to_index["title"].tolist()


# Generate text embeddings for all titles
faiss_title_embedding = model.encode(titles_list, convert_to_numpy=True)

# # Normalize the embeddings for cosine similarity (FAISS works better with normalized vectors)
content_encoded_normalized = faiss_title_embedding.copy()
faiss.normalize_L2(content_encoded_normalized)

# Create a NumPy array containing the document IDs
id_index = np.array(pdf_to_index["id"].tolist())  # IDs des documents

# # Create a FAISS index using an IndexFlatIP (inner product for similarity search)
index_content = faiss.IndexIDMap(faiss.IndexFlatIP(content_encoded_normalized.shape[1]))

# Add the normalized embeddings and corresponding document IDs to the FAISS index
index_content.add_with_ids(content_encoded_normalized, id_index)

#
print(f"Nombre de documents indexés : {index_content.ntotal}")
```

```python
def search_content(query, pdf_to_index, k=10):
     """
    Searches for the top k most similar documents in FAISS.

    Parameters:
    - query : (str) The search text
    - pdf_to_index : (DataFrame) The DataFrame containing indexed documents
    - k : (int) The number of search results to return

    Returns:
    - A DataFrame of the most similar documents and their similarity scores.
    """
    # Encode the query into a numerical embedding
     query_vector = model.encode([query], convert_to_numpy=True)


    # Normalize the query embedding for cosine similarity comparison
     faiss.normalize_L2(query_vector)

     # Perform the search and retrieve the top k most similar documents
     distances, ids = index_content.search(query_vector, k)

    # Retrieve the matching documents from the original DataFrame
     results = pdf_to_index[pdf_to_index["id"].isin(ids[0])].copy()


    # Add similarity scores to the DataFrame
     results["similarity"] = distances[0]

     return results
```

```python
# Define a test query
query_test = "animal"


# Retrieve the top 5 most similar articles related to "animal"
top_k_results = search_content(query_test, pdf_to_index, k=5)


# Display the results with relevant columns
print("\n Résultats de la recherche :")
print(top_k_results[["id", "title", "similarity"]])
```

Question Answering with Hugging Face Model

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

# Initialisation of the model GPT-2 and the tokenizer
model_id = "gpt2"  # Modèle pré-entraîné
tokenizer = AutoTokenizer.from_pretrained(model_id)
lm_model = AutoModelForCausalLM.from_pretrained(model_id)

# Create pipeline for text generation
pipe = pipeline(
    "text-generation",
    model=lm_model,
    tokenizer=tokenizer,
    max_new_tokens=150,  # Maximum number of tokens generated
    device_map="auto",  # Automatic use of GPU if available
)

# Search ChromaDB for context
question = "What’s the latest news on space development?"  # user's question
results = search_chroma(question, top_k=3)  # We retrieve the 3 most relevant articles

# Building the prompt with the retrieved documents
context = " ".join(results["documents"][0])  # We retrieve the 3 most relevant articles

prompt_template = f"Relevant context: {context}\n\nUser's question: {question}\nAnswer:"

# Generating the response with GPT-2
lm_response = pipe(prompt_template)


# Dispaly generated text
print("\n Réponse générée :")
print(lm_response[0]["generated_text"])
```
