Use NumPy to generate a synthetic dataset representing average monthly temperatures (in degrees Celsius) for 12 months across 10 different cities. The temperatures should range from -5 to 35 degrees.

Convert this NumPy array into a Pandas DataFrame, adding city names as index and months as columns.

```python
temp_data = np.random.uniform(low=-5, high=35, size=(10, 12))
temp_data
```

```python
city = ["Paris", "London", "Berlin", "Roma", "Madrid", "Luxembourg", "Geneva",
        "Bruxel", "Athena", "Lisbon"]

temp_cities = [[random.randint(-5, 35) for _ in range(12)] for _ in range(10)]
temp_cities
```

```python
df = pd.DataFrame(temp_cities, index=city, columns=["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
df.head()
```

**2. Data Analysis:**

**Hint 1:** Calculate the annual average temperature using DataFrame.mean(axis).


**Hint 2:** Find the city with the highest and lowest average temperature using idxmax() and idxmin() methods.

* Calculate the annual average temperature for each city.

* Identify the city with the highest and lowest average temperature for the year.

```python
# Loop through the 10 cities
for i in range(10):
    # Calculate the average temperature for city i (mean of all months)
    average_i = df.iloc[i].mean()

    # Print the result with the city name and rounded temperature
    print(f"The average temperature in {df.index[i]} is {round(average_i)} degrees")
```

**3. Data Visualization:**

```python
# Create the figure and axes for the plot
fig, ax1 = plt.subplots(figsize=(32, 8))

# Loop to plot the temperature variations for each city
for i in range (10):
  ax1.plot(df.columns, df.iloc[i], label=df.index[i])

# Add labels for the axes
ax1.set_xlabel("Month")
ax1.set_ylabel("Temperature (Â°C)")
ax1.legend()
ax1.set_title("Fluctuation of monthly temperatures in European cities")

# Add a grid to improve readability
ax1.grid(True)
plt.show()
```
