***Exercice 2***

To convert each of following unstructured data sources into structured data we will use :

1. A series of blog posts about travel experiences

    * Method: Natural Language Processing (NLP)   
    * NLP allows segmenting text into categories, extracting key entities (such
     as locations, dates, and sentiments), and structuring the information into a dictionary or a table (DataFrame).


2. Audio recordings of customer service calls

    * Method: Automatic Speech Recognition (Speech-to-Text)
    * This technique converts audio into text, which can then be analyzed to identify keywords, emotions, and call trends. The text can be structured into a table with columns such as "Call ID", "Customer", "Sentiment", and "Call Reason".


3.  Handwritten notes from a brainstorming session

    * Method: Optical Character Recognition (OCR)
    * OCR (such as Tesseract) automatically extracts handwritten text and converts it into digital text. The extracted information can then be structured in a table with columns like "Idea", "Category", and "Priority".


4. A video tutorial on cooking, we can use :   

        * Keyframe extraction (OpenCV) wich captures important moments of the
           video.
        * We can use other tools for detect objects such as ingredients ant
           utensils.
        * We can also use a transcription tools (Speech-to-text).
        * Finally, we orgnize all these elements into a dataset with columns  
          like "Ingredients", "Action", "duration"...
