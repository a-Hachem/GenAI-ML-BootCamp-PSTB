In the following, we will see two different methods to create the "Title" column from the "Name" column.

**Method 1**: Using the split() function



```python
# This function extracts the title from the "Name" column using a simple text split operation.
def extract_title(name):
    words = name.split()  # Split the name into words
    for word in words:
        if "." in word:  # Check if the word contains a period (which indicates a title)
            return word.replace(".", "")  # Remove the period and return the title
    return "Unknown"  # If no title is found (very rare cases)

# Apply the function to create the "Title" column
df1["Title"] = df1["Name"].apply(extract_title)

# Display unique titles
print(df1["Title"].unique())
```

**Method 2**: Using "Regular Expressions"

```python
# Using regex to extract the title efficiently
df1["Title"] = df1["Name"].str.extract(r' ([A-Za-z]+)\.')

# Display unique titles
print(df1["Title"].unique())
```

 In this section, we will detect and handle outliers in the "Age" and "Fare" columns.

  We will use two common methods for detecting outliers:
1.  **Interquartile Range (IQR)**: Identifies outliers based on the range between Q1 (25th percentile) and Q3 (75th percentile).

2.  **Z-score**: Measures how many standard deviations a value is from the mean.

After detecting outliers, we will decide whether to cap, transform, or remove them.


```python
import numpy as np
from scipy import stats

# Function to detect outliers using IQR
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)  # 25th percentile
    Q3 = df[column].quantile(0.75)  # 75th percentile
    IQR = Q3 - Q1  # Interquartile range
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

# Function to detect outliers using Z-score
def detect_outliers_zscore(df, column, threshold=3):
    z_scores = np.abs(stats.zscore(df[column]))
    return df[z_scores > threshold]

# Detecting outliers in "Fare" and "Age"
outliers_fare_iqr = detect_outliers_iqr(df1, "Fare")
outliers_age_iqr = detect_outliers_iqr(df1, "Age")

outliers_fare_z = detect_outliers_zscore(df1, "Fare")
outliers_age_z = detect_outliers_zscore(df1, "Age")

#  Displaying the number of detected outliers
print(f"- Outliers in 'Fare' (IQR method): {len(outliers_fare_iqr)}")
print(f"- Outliers in 'Age' (IQR method): {len(outliers_age_iqr)}")
print(f"- Outliers in 'Fare' (Z-score method): {len(outliers_fare_z)}")
print(f"- Outliers in 'Age' (Z-score method): {len(outliers_age_z)}")
```

The choice of the method for detecting outliers depends on the distribution of the variable. If the distribution is skewed, the IQR method is more appropriate as it does not assume normality. However, if the distribution is more normal, the Z-score method is preferable, as it accurately identifies significant deviations from the mean without being overly sensitive to natural variations.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Tracer la distribution de "Fare" et "Age" avant traitement des outliers
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Distribution de "Fare"
sns.histplot(df1["Fare"], bins=50, kde=True, ax=axes[0], color="blue")
axes[0].axvline(np.mean(df1["Fare"]), color='red', linestyle='dashed', linewidth=2, label="Mean")
axes[0].set_title("Distribution of Fare (Before Outlier Handling)")
axes[0].set_xlabel("Fare")
axes[0].set_ylabel("Count")
axes[0].legend()

# Distribution de "Age"
sns.histplot(df1["Age"], bins=50, kde=True, ax=axes[1], color="green")
axes[1].axvline(np.mean(df1["Age"]), color='red', linestyle='dashed', linewidth=2, label="Mean")
axes[1].set_title("Distribution of Age (Before Outlier Handling)")
axes[1].set_xlabel("Age")
axes[1].set_ylabel("Count")
axes[1].legend()

# Afficher la figure
plt.tight_layout()
plt.show()
```

Since "Fare" is skewed, the IQR method is more suitable as it does not assume normality and effectively captures extreme price variations. However, for "Age", which is more normally distributed, the Z-score method is a better choice as it accurately identifies values that deviate significantly from the mean without being overly sensitive to natural variations.



```python
import numpy as np

# 1. Transformation logarithmique pour "Fare" (ajout de 1 pour viter log(0))
df1["Fare_log"] = np.log1p(df1["Fare"])

# 2. Capping des valeurs extrmes pour "Age"
Q1_Age = df1["Age"].quantile(0.05)  # 5e percentile
Q3_Age = df1["Age"].quantile(0.95)  # 95e percentile

df1["Age_capped"] = df1["Age"].clip(Q1_Age, Q3_Age)  # Remplacement des valeurs extrmes

# Visualisation des nouvelles distributions
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Distribution de "Fare" aprs transformation logarithmique
sns.histplot(df1["Fare_log"], bins=50, kde=True, ax=axes[0], color="blue")
axes[0].axvline(np.mean(df1["Fare_log"]), color='red', linestyle='dashed', linewidth=2, label="Mean")
axes[0].set_title("Distribution of Fare (After Log Transformation)")
axes[0].set_xlabel("Log(Fare)")
axes[0].set_ylabel("Count")
axes[0].legend()

# Distribution de "Age" aprs capping
sns.histplot(df1["Age_capped"], bins=50, kde=True, ax=axes[1], color="green")
axes[1].axvline(np.mean(df1["Age_capped"]), color='red', linestyle='dashed', linewidth=2, label="Mean")
axes[1].set_title("Distribution of Age (After Capping)")
axes[1].set_xlabel("Age")
axes[1].set_ylabel("Count")
axes[1].legend()

plt.tight_layout()
plt.show()
```

After applying the log transformation to 'Fare', the distribution shifted slightly to the right, reducing the impact of extreme values. However, some outliers are still present. Similarly, after capping 'Age', the distribution became more balanced, but a few extreme values remain. Another option is to apply an additional transformation, like the square root or Box-Cox transformation, to better normalize the data.
